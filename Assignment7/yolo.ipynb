```python
from ultralytics import YOLO

# Load a pretrained YOLOv8 model
model = YOLO('yolov8n.pt')  # 'n' stands for nano; you can choose 's', 'm', 'l', 'x'
```

```python
imagepath = '/content/image2.jpeg'
```

```python
results = model(imagepath)

# Display results
for result in results:
  result.show()
```

```python
for result in results:
    boxes = result.boxes.xyxy  # Box coordinates
    labels = result.boxes.cls  # Class labels
    scores = result.boxes.conf  # Confidence scores
    print(boxes, labels, scores)
```
